auto correlation analysis는 보통 시계열 분석에서 사용됨
-->시계열 분석이란, 어떤 현상에 대하여 과거에서부터 현재까지의 시간의 흐름에 따라 기록된 데이터를 바탕으로 미래의 변화에 대한 추세를 분석하는 방법
