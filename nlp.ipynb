```python
from nltk.tokenize import word_tokenize
nltk.download('punkt') #punkt-->puctuation에 대한 데이터
words = word_tokenize(text)
#word_tokenize function쓰면 punctuation까지 나눠줌
```

```python
from nltk.tokenize import RegexpTokenizer
retokenize = RegexpTokenizer("[\w]+")
words = retokenize.tokenize(text)
#retokenize(regular expression based tokenize)는 puctuation을 아예 제거
```

### Normalization  
**Stemming** 어간 추출 대충의 패턴 규칙으로 어미를 잘라내는 것 (사전에 없는 어간 나올 수 있음)

**Lemmatization** 표제어(기본 사전형) 추출. -->잘라내는 것만 하지 않고 원형으로 복귀


```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
[stemmer.stem(w) for w in words]
#words는 regular expression based tokenize로 위에서 자름
#각각의 단어에 대해서 for loop
#stemming은 패턴으로 자르다보니 사전에 없는 것들이 나옴
```

```python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
[stemmer.stem(w) for w in words]
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
[lemmatizer.lemmatize(w) for w in words]
#정교한 tokenizer, 단어에 대한 정보 필요해서 워드넷 사용
```

### Stopword
너무 많이 쓰이는 단어 제거할때

```python
from nltk.corpus import stopwords  
nltk.download('stopwords') #nltk에서 제공
print(words)#이건 원래 words
words = [w for w in words if not w in stopwords.words('english')] 
#stop words에 단어들이 있는데 하나하나씩 받아서 검사를 한다-->stopwords에 들어있지 않으면 취하라
print(words)#stopwords 거른 words
```

### Collocation, Concordance
연어,

```python
nltk.download('gutenberg')
#gutenberg-->글에 대해서 저작권이 지난 classic들을 무료로 제공
text = nltk.corpus.gutenberg.raw('austen-emma.txt')
#emma라는 소설을 가져옴-->text는 string으로 되어있음
words = retokenize.tokenize(text)
#words는 list
```

```python
nltk.Text(words).collocations()  
# default: (num=20, window_size=2)-->parameter 값 default
#-->20개 나오고 연속되는 단어 갯수 2개
#window size는 연속되는 단어 갯수
#collocation-->두개가 연속해서 나오는 단어들
```

```python
nltk.Text(words).concordance('Emma', 79, 10)
#concordance-->용매-->corpus 전체에서 Emma라는 단어가 등장하는 앞뒤 context를 같이 보여줌
#-->총 10개를 뽑아라
#79는 앞뒤 context의 character 숫자의 갯수
```

```python
nltk.Text(words).dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"])
#dispersion_plot-->Emma라는게 등장하는게 어딘지 표현
#다양한 단어들이 어떻게 분포되어있는지 알 수 있다
```

```python
# Distributional similarity: 
# find other words which appear in the same contexts as the specified word; 
# list most similar words first.
nltk.Text(words).similar("Emma")

#similar는 양쪽 옆으로 Emma하고 비슷하게 나올 단어를 찾아라-->대부분 대명사
```

```python
# Find contexts where the specified words appear; list most frequent common contexts first.
#specified words가 나타나는 context를 말하라
#_에 Emma하고 she하고 같이 등장할 가능성이 높다
nltk.Text(words).common_contexts(["Emma", "she"])
```

```python
'''       Basic Regular Expression Meta-Characters, Including Wildcards, Ranges and Closures

.	        Wildcard, matches any character
^abc	    Matches some pattern abc at the start of a string
abc$	    Matches some pattern abc at the end of a string
[abc]	    Matches one of a set of characters
[^abc]    Matches anything but a set of characters
[A-Z0-9]	Matches one of a range of characters
ed|ing|s	Matches one of the specified strings (disjunction)
*	        Zero or more of previous item, e.g. a*, [a-z]* (also known as Kleene Closure)
+	        One or more of previous item, e.g. a+, [a-z]+
?	        Zero or one of the previous item (i.e. optional), e.g. a?, [a-z]?
{n}	      Exactly n repeats where n is a non-negative integer
{n,}	    At least n repeats
{,n}	    No more than n repeats
{m,n}	    At least m and no more than n repeats
a(b|c)+	  Parentheses that indicate the scope of the operators
(...)     Matches whatever regular expression is inside the parentheses
\d
Matches any decimal digit; this is equivalent to the class [0-9].
\D
Matches any non-digit character; this is equivalent to the class [^0-9].
\s
Matches any whitespace character; this is equivalent to the class [ \t\n\r\f\v].
\S
Matches any non-whitespace character; this is equivalent to the class [^ \t\n\r\f\v].
\w
Matches any alphanumeric character; this is equivalent to the class [a-zA-Z0-9_].
\W
Matches any non-alphanumeric character; this is equivalent to the class [^a-zA-Z0-9_].

'''
```

```python
nltk.download('words')
engdict = nltk.corpus.words.words('en')
#list로 영어사전이 들어옴
```

```python
result = [w for w in engdict if re.search('ed$', w)]
#engdict에 있는 각각의 item을 w로 받아서 그 갯수만큼  for loop를 돈다
#-->각 loop마다 if가 적용되는데 w로 들어와서 'ed$'를 찾아서 w로 list해라-->ed가 있으면 통과
#ed$-->string의 끝에 ed가 있으면
#^abc-->string의 시작에서 abc가 있으면 
print(result)
```

```python
result = [w for w in engdict if re.search('^..j..t..$', w)]
# ^..j..t..$-->.은 어떤 character가 와도 됨-->_ _j_ _t_ _이런 단어 찾아라
print(result)
```

```python
result = [w for w in engdict if re.search('^[ghi][mno][jlk][def]$', w)]
#대괄호는 그중에 하나를 만족시키는-->or
#^[ghi][mno][jlk][def]$-->4개의 character로 되어 있을 것
print(result)
```

```python
result = [w for w in engdict if re.search('^[ah]+$', w)][:10]
#+는 한번 이상
#^[ah]+$-->a혹은 h가 한개 이상 나오고 끝나는 것
print(result)
```

```python
nltk.download('treebank')
#treebank-->wallstreetjournal에 있는 단어-->현실에 가까운 단어 corpus
wsj = nltk.corpus.treebank.words()

result = [w for w in wsj if re.search('(ed|ing)$', w)]
#(ed|ing)-->ed혹은 ing로 끝나는 단어

# result = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
#^[0-9]+\.[0-9]+$-->숫자가 하나 이상 나오고 .이 나오고 숫자가 하나 이상 나오는 것
#\.은 실제 점을 뜻함
#전화번호부, 이메일의 형태에 사용가능

# result = [w for w in wsj if re.search('^[A-Z]+\$$', w)]
#\$-->실제 달러
#[A-Z]+\$$-->어떠한 대문자가 하나 이상나오고 $표시로 끝나는 것

# result = [w for w in wsj if re.search('^[0-9]{4}$', w)]
#{}-->{}앞에 있는 것이 몇번 나오는가
#[0-9]{4}$-->4자리 숫자만 뽑아내는 경우
# result = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
#숫자가 하나 이상 나오고, - 나오고, 소문자가 3개에서 5개사이에서 나오고 끝나는 것
# result = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
#소문자가 5개이상, -나오고,소문자가 2개에서 3개-소문자가 6개 이하
result = sorted(set(result))
#sort의 작업을 해줌-->알파벳순
print(result[:10])
```

```python
import os
url = "https://raw.githubusercontent.com/hsnam95/my/main/friends_season01_script.txt"
os.system("curl " + url + " > friends_season01_script.txt")
#파일을 서버에 가져옴
# read a text file in the server-->서버의  text file읽기
file = open("friends_season01_script.txt")
text = file.read()
file.close()
text
#text는 string
#len은 character의 갯수
```

```python
len(text)
```

```python
pattern = '(?<=: ).+(?=[\.|\?|\!])'
#(?<=).(?=) 앞조건 개별 조건 뒤조건
#캐릭터가 나오고 :이고 punctuation으로 끝나는 것
#:이 앞에 나오고 하나 이상의 캐릭터가 나오고 \.,\?,\!든지
#혼동 방지를 위한 \ 사용
sent = re.findall(pattern, text)
#pattern을 만족시키는 것을 sent에 list
print(sent)
text = '\n'.join(sent)
#list로 만든 것을 재결합
#줄바꾼거 다 붙여라
```
